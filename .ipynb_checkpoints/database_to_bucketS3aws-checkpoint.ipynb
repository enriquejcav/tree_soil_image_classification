```python
import boto3
import os.path
import sagemaker
from typing import Callable
from sagemaker.session import Session
```

## Uploading database to S3 BUCKET
**upload_folder_to_s3:** Send the database to each specific folder in the bucket, as in the project's master directory

helped me: https://stackoverflow.com/questions/25380774/upload-a-directory-to-s3-with-boto

```python
def upload_folder_to_s3(s3bucket:Callable, inputDir:str, s3Path:str):
        print("Uploading results to s3 initiated...")
        print("Local Source:",inputDir)
        os.system("ls -ltR " + inputDir)

        print("Dest  S3path:",s3Path)

        try:
            for path, subdirs, files in os.walk(inputDir):
                for file in files:
                    dest_path = path.replace(inputDir,"")
                    __s3file = os.path.normpath(s3Path + '/' + dest_path + '/' + file)
                    __local_file = os.path.join(path, file)
                    print("upload : ", __local_file, " to Target: ", __s3file, end="")
                    s3bucket.upload_file(__local_file, __s3file)
                    print(" ...Success")
        except Exception as e:
            print(" ... Failed!! Quitting Upload!!")
            print(e)
            raise e
```

```python
bucket_name = 'pixforce-awsbucket-s3' # <--- CHANGE THIS VARIABLE TO A UNIQUE NAME FOR YOUR BUCKET
my_region = boto3.session.Session().region_name # set the region of the instance
folders_to_copy =["Data_in_Classes/","Data_fft_in_Classes/"]
print('Region: {}, Bucket Name: {}, Folders to copy: {}'.format(my_region,bucket_name,folders_to_copy))
s3 = boto3.resource('s3', region_name=my_region)
s3bucket = s3.Bucket(bucket_name)
for folder_to_copy in folders_to_copy:
    upload_folder_to_s3(s3bucket, folder_to_copy, folder_to_copy)
    print( folder_to_copy + "Uploud Sucesss")
```
