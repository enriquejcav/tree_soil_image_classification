# Structuring the database

A Jupyter Notebook to organize dataset into a useful structure


## Importations to use

```python
# organize dataset into a useful structure
import cv2
import numpy as  np
from os import makedirs
from os import listdir
from os.path import isfile, join
from shutil import copyfile
from random import seed
from random import random
from matplotlib import pyplot
from matplotlib.image import imread
from numbers import Number
```

## Creating Directorys and Subdirector
**data_in_classes_function:** Creates folders to allocate the database in a structured way.

```python
def data_in_classes_function(dataset_home:list,subdirs:list):
    #Creating directory structure
    for elem_dataset_home in dataset_home:
        for subdir in subdirs:
            # create label subdirectories
            labeldirs = ['tree/', 'soil/']
            for labldir in labeldirs:
                newdir = elem_dataset_home + subdir + labldir
                makedirs(newdir, exist_ok=True)
```

```python
dataset_home = ['Data_in_Classes/', 'Data_fft_in_Classes/']
subdirs = ['train/', 'val/', 'test/']
data_in_classes_function(dataset_home,subdirs)
```

## Set Fourrier Transform Function in a Image
**fourrier_transform:** Calculate the image fourier transform

```python
def fourrier_transform(src:str): 
    img_to_do_fft = cv2.imread(src,0)
    dft = cv2.dft(np.float32(img_to_do_fft), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
    return magnitude_spectrum
```

## Functions to copy files to train/val/test  folders
**select_directory_to_copy:** Select directory for copy distributing in train/val/test folders according to the proportion ratio.

**copying_image:** Copies the image randomly to each folder

**copying_fft** Copies the Fourrier Transform image randomly to each folder



*Important report: 70% to train, 15% to validation and 15% to test*

```python
def select_directory_to_copy(val_ratio_inf:Number, val_ratio_sup:Number) -> str:
    #To know whats train/val/test
    if random() < val_ratio_sup and random() > val_ratio_inf:
        alocation_dir = 'val/'
    elif random() < val_ratio_inf:
        alocation_dir = 'test/'
    else:
        alocation_dir = 'train/'
    return alocation_dir

def copying_image(dataset_home:str, directory:str, val_ratio_inf:Number, val_ratio_sup:Number):
    # seed random number generator
    seed(1)
    for file in listdir(directory):
        src = directory + '/' + file
        alocation_dir = select_directory_to_copy(val_ratio_inf,val_ratio_sup)
        #Alocation
        if file.startswith('tree',5):
            dst = dataset_home + alocation_dir + 'tree/'  + file
            copyfile(src, dst)
        elif file.startswith('soil',5):
            dst = dataset_home + alocation_dir + 'soil/'  + file
            copyfile(src, dst)
        
def copying_fft(dataset_home:str, directory:str, val_ratio_inf:Number, val_ratio_sup:Number):
    # seed random number generator
    seed(1)
    for file in listdir(directory):
        src = directory + '/' + file
        alocation_dir = select_directory_to_copy(val_ratio_inf,val_ratio_sup)
        #Alocation
        if file.startswith('tree',5):
            magnitude_spectrum = fourrier_transform(src)
            dst = dataset_home + alocation_dir + 'tree/fft'  + file
            cv2.imwrite(dst,magnitude_spectrum)
        elif file.startswith('soil',5):
            magnitude_spectrum = fourrier_transform(src)
            dst = dataset_home + alocation_dir + 'soil/fft'  + file
            cv2.imwrite(dst,magnitude_spectrum)
```

```python
# define ratio of pictures to use for validation
val_ratio_inf = 0.15
val_ratio_sup = 0.3
# copy training dataset images into subdirectories
directory ='Data/' 

copying_image(dataset_home[0],directory,val_ratio_inf,val_ratio_sup)
print("Images Successfully Copied")
copying_fft(dataset_home[1],directory,val_ratio_inf,val_ratio_sup)
print("FFT Images Successfully Copied")
```

## Number of files
A simple implementation to see how many files have in each directory and subdirectory

```python
def number_files(dataset):
    number_files = {}
    for elem in listdir(dataset): 
        path = dataset + elem + '/'
        for subelem in listdir(path):
            subpath = path + subelem + '/'
            number_files[elem + '_' + subelem] = len([f for f in listdir(subpath) if isfile(join(subpath, f))])
    print('Dataset: {}\n N° examples {}\n'.format(dataset,number_files))
```

```python
number_files(dataset_home[0])
number_files(dataset_home[1])
```
